```
import matplotlib.pyplot as plt
# x-axis values
day = [1, 2, 3, 4, 5,6,7]
# Y-axis values
temp = [45,49,43,42,47,48,43]
#Function to plot graph
plt.plot(day, temp,color='red',linestyle='dotted', marker='o', markerfacecolor="blue" ) # marker o for circles and . is used for points
plt.title("Weather Temperature graph")
plt.ylabel("Temperature")
plt.xlabel("Days")
# function to show the plot
plt.show()
```
